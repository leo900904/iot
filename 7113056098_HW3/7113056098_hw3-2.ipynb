Write a Python script to simulate and visualize two classes of data points in a 2D space. Follow these steps:

Generate 600 random data points (x1, x2) from a normal distribution with a mean of 0 and a variance of 10.
Calculate the Euclidean distance of each point from the origin.
Classify the points: assign a label Y=0 to points within a distance of 10 from the origin, and Y=1 to the points outside this radius.
Plot the data points using matplotlib:
Use blue circles for points labeled Y=0.
Use red squares for points labeled Y=1.
Include labels for the axes, a title, a legend, and a grid."

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 600 random data points (x1, x2) from a normal distribution
n_points = 600
mean = 0
variance = 10
x1 = np.random.normal(mean, np.sqrt(variance), n_points)
x2 = np.random.normal(mean, np.sqrt(variance), n_points)

# Step 2: Calculate the Euclidean distance from the origin
distances = np.sqrt(x1**2 + x2**2)

# Step 3: Classify points based on their distance from the origin
radius_threshold = 10
labels = (distances > radius_threshold).astype(int)

# Step 4: Plot the data points
plt.figure(figsize=(10, 7))

# Plot points with label Y=0 (blue circles)
plt.scatter(x1[labels == 0], x2[labels == 0], color='blue', marker='o', label='Y=0 (Inside Radius)')

# Plot points with label Y=1 (red squares)
plt.scatter(x1[labels == 1], x2[labels == 1], color='red', marker='s', label='Y=1 (Outside Radius)')

# Add plot features
plt.title('2D Data Point Classification')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Ensure aspect ratio is equal for better visualization

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Generate 600 random data points centered at (0, 0) with variance 10
n_points = 600
mean = 0
variance = 10
x1 = np.random.normal(mean, np.sqrt(variance), n_points)
x2 = np.random.normal(mean, np.sqrt(variance), n_points)

# Calculate Euclidean distance from the origin
distances = np.sqrt(x1**2 + x2**2)

# Classify points: Y=0 if distance < 4, Y=1 otherwise
labels = (distances >= 4).astype(int)

# Step 1: Scatter plot (2D)
plt.figure(figsize=(10, 7))
plt.scatter(x1[labels == 0], x2[labels == 0], color='blue', marker='o', label='Y=0 (Inside radius)')
plt.scatter(x1[labels == 1], x2[labels == 1], color='red', marker='s', label='Y=1 (Outside radius)')
plt.title('Step 1: 2D Data Points with Labels')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()

# Step 2: Define x3 as a Gaussian function
def gaussian(x1, x2, mu=0, sigma=2):
    return np.exp(-((x1 - mu)**2 + (x2 - mu)**2) / (2 * sigma**2))

x3 = gaussian(x1, x2)

# Step 2: 3D Scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot points with Y=0 and Y=1 in different colors
ax.scatter(x1[labels == 0], x2[labels == 0], x3[labels == 0], color='blue', marker='o', label='Y=0')
ax.scatter(x1[labels == 1], x2[labels == 1], x3[labels == 1], color='red', marker='s', label='Y=1')

ax.set_title('Step 2: 3D Data Points (x1, x2, x3)')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.legend()
plt.show()

# Step 3: Plot a linear hyperplane (e.g., z = 0.5 * x1 + 0.5 * x2)
x1_range = np.linspace(min(x1), max(x1), 50)
x2_range = np.linspace(min(x2), max(x2), 50)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
x3_hyperplane = 0.5 * x1_grid + 0.5 * x2_grid

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the points
ax.scatter(x1[labels == 0], x2[labels == 0], x3[labels == 0], color='blue', marker='o', label='Y=0')
ax.scatter(x1[labels == 1], x2[labels == 1], x3[labels == 1], color='red', marker='s', label='Y=1')

# Plot the hyperplane
ax.plot_surface(x1_grid, x2_grid, x3_hyperplane, color='lightblue', alpha=0.5)

ax.set_title('Step 3: 3D Data Points with Linear Hyperplane')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LogisticRegression

# Step 1: Generate 600 random data points centered at (0, 0) with variance 10
n_points = 600
mean = 0
variance = 10
x1 = np.random.normal(mean, np.sqrt(variance), n_points)
x2 = np.random.normal(mean, np.sqrt(variance), n_points)

# Calculate Euclidean distance from the origin
distances = np.sqrt(x1**2 + x2**2)

# Classify points: Y=0 if distance < 4, Y=1 otherwise
labels = (distances >= 4).astype(int)

# Step 1: 2D Scatter plot
plt.figure(figsize=(10, 7))
plt.scatter(x1[labels == 0], x2[labels == 0], color='blue', marker='o', label='Y=0 (Inside radius)')
plt.scatter(x1[labels == 1], x2[labels == 1], color='red', marker='s', label='Y=1 (Outside radius)')
plt.title('Step 1: 2D Data Points with Labels')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()

# Step 2: Define x3 as a Gaussian function
def gaussian(x1, x2, mu=0, sigma=2):
    return np.exp(-((x1 - mu)**2 + (x2 - mu)**2) / (2 * sigma**2))

x3 = gaussian(x1, x2)

# Step 2: 3D Scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x1[labels == 0], x2[labels == 0], x3[labels == 0], color='blue', marker='o', label='Y=0')
ax.scatter(x1[labels == 1], x2[labels == 1], x3[labels == 1], color='red', marker='s', label='Y=1')

ax.set_title('Step 2: 3D Data Points (x1, x2, x3)')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.legend()
plt.show()

# Step 3 & 4: Fit a linear hyperplane and plot it in 3D
# Logistic Regression to find the best linear separator
X = np.column_stack((x1, x2))
clf = LogisticRegression().fit(X, labels)

# Generate hyperplane values
x1_range = np.linspace(min(x1), max(x1), 50)
x2_range = np.linspace(min(x2), max(x2), 50)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

# Hyperplane equation: w1*x1 + w2*x2 + b = 0
# Solve for x3: x3 = -(w1*x1 + w2*x2 + b) / coeff (fake separator for 3D)
z_hyperplane = -(clf.coef_[0][0] * x1_grid + clf.coef_[0][1] * x2_grid + clf.intercept_)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x1[labels == 0], x2[labels == 0], x3[labels == 0], color='blue', marker='o', label='Y=0')
ax.scatter(x1[labels == 1], x2[labels == 1], x3[labels == 1], color='red', marker='s', label='Y=1')

ax.plot_surface(x1_grid, x2_grid, z_hyperplane, color='lightblue', alpha=0.5)
ax.legend()
plt.show()

```
